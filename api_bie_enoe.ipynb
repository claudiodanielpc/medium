##Uso de la API del Banco de Información Económica del INEGI
#Población ocupada trimestral (ENOE)

```python
#Librerías
import requests
import pandas as pd
```

```python
#url con token y respuesta
#Datos de personal ocupado
url = "https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/289245/es/0700/false/BIE/2.0/a004f0de-f1d3-0adf-1bfb-81bac98fa7b5?type=json"
respuesta = requests.get(url).json()
respuesta.keys()
respuesta
```

```python
#Guardamos únicamente los elementos de Series
result=respuesta["Series"]
```

```python
#checar tipo
type(result)
```

```python
#Transformar a dataframe
df=pd.DataFrame.from_records(result)
df
```

```python
#Obtener datos de OBSERVATIONS
df = df["OBSERVATIONS"].tolist()
#Hacer dataframe únicamente con la lista OBSERVATIONS
df=pd.DataFrame.from_records(df[0])
#Columnas a minúsculas
df.columns= df.columns.str.lower()
#Selección de variables
df= df[["time_period", "obs_value"]]
#Desplegar dataframe
df
```
