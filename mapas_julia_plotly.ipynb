```julia
using PlotlyJS, HTTP, JSON, DataFrames, CSV
```

```julia
#Obtener información de inmuebles irregulares de la CDMX
url="https://datos.cdmx.gob.mx/dataset/0e60be38-58b4-4f7e-b9f5-0c326d196c75/resource/1fac6305-35e3-43c5-b3ba-db814ec6cc8c/download/desarrollos_irregulares.csv"
#Descargar el archivo
download(url, "desarrollos_irregulares.csv")
#Leer el archivo
df = CSV.read("desarrollos_irregulares.csv", DataFrame)
#Solo mostrar los 3 primeros registros
first(df, 3)
```

```julia
puntos=scattermapbox(;lat=df[!,:latitud], lon=df[!,:longitud], mode="markers",text=df[!,:ubicacion],
marker=attr(;size=10, color="#756bb1"))


    layout = Layout(;title="Desarrollo irregulares en la alcaldía Benito Juárez",
    
    geo_resolution=50,
    width=1000, height=700,
    mapbox=attr(
        style="open-street-map",  # Estilo del mapa
        center=attr(lat=19.3804904, lon=-99.1615416),  # Coordenadas de la alcaldía Benito Juárez
        zoom=13  # Zoom
    ))
#Mostrar mapa
plot(puntos,layout)
```
